We are using miniconda3 with Python 3.10.9, TensorFlow 2.10.0



We will need these 2 below libraries to create our training numpy arrays, where the X comprises random numbers, Y is the multiples of X by 2.

```python
import random
import numpy as np
```

We will divide our X values by 10 to make it easier to handle by Keras further down the line.

```python


X = np.random.randint(0, 10, 10000)
X = X/10
Y = np.multiply(X,2)
#X is the input array and Y is the output array

```

here's the keras model - we're using the very much defaults. We're running 100 epochs, which is a bit overkill here. As always, depends what we are trying to achieve. Could have used 75 - see branches. At 50 epochs, we were stil getting quite a lot of inconsistency in the fractional part of the predicted numbers.

```python

from keras.models import Sequential
from keras.layers import Activation, Dense

model=Sequential()
model.add(Dense(1,activation='relu',input_shape=(1,)))
model.add(Dense(1,activation='relu'))
model.compile(loss='MSE', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, batch_size=50, validation_split=0.3, epochs=100, verbose=1,shuffle=1)
```

```python

test_array=np.array([4,27,100,121,9,12,8,79,2,2151,1000000])
prediction_array = model.predict([test_array])

for i in range(len(test_array)):
    print("here's our: \t test: {} \t predicted: {}".format(test_array[i],prediction_array[i][0]))
```
